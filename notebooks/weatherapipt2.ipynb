```python
def get_astronomy(city):
    url = f"{baseurl}astronomy.json?key={jayson_key}&q={city}&dt={today}"
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        location = data['location']['name']
        sunrise = data['astronomy']['astro']['sunrise']
        sunset = data['astronomy']['astro']['sunset']
        moonrise = data['astronomy']['astro']['moonrise']
        moonset = data['astronomy']['astro']['moonset']
        moonphase = data['astronomy']['astro']['moon_phase']

        print(f"Location is {location}")
        print(f"Sunrise is: {sunrise}")
        print(f"Sunset is: {sunset}")
        print(f"Moonrise is: {moonrise}")
        print(f"Moonset is: {moonset}")
        print(f"Moon Phase is: {moonphase}")
    else:
        print("Error fetching data")
        
```
